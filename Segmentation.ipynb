The function main() requires following inputs:

function: the name of the selected algorithm, such as "CV", "RSF", and so on. The code of LGMLS and ILLS is hid at present.

T:  the threshold in some threshold-based algorithms.
Gsig: sigma in the Gaussian kernel function in PSF.
filepath: the path of the input image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import ndimage
import skimage
from skimage import util
from time import *

def GaussKernel(sig):
    kernel = np.ones((4*sig+1,4*sig+1),np.float64)
    for i in range(4*sig+1):
        for j in range(4*sig+1):
            kernel[i,j]=1/(2*math.pi*sig**2)*math.exp(-((i-2*sig)**2+(j-2*sig)**2)/(2*sig**2))
    return kernel/kernel.sum()

def Gmax(A,size):
    pad=(int)((size-1)/2)
    A_ori=A
    A=np.insert(A,0,np.zeros((pad,A.shape[1])),axis=0)
    A=np.insert(A,0,np.zeros((pad,A.shape[0])),axis=1)
    A=np.append(A,np.zeros((pad,A.shape[1])),axis=0)
    A=np.append(A,np.zeros((A.shape[0],pad)),axis=1)
    B=np.zeros_like(A_ori)
    for i in range(A_ori.shape[0]):
        for j in range(A_ori.shape[1]):
            B[i,j]=A[i:i+size,j:j+size].max()
    return B

def level_set(LSF,img,mu,nu,lambda1,lambda2,dr,sig,epison,step,kx,function):
    kernel=GaussKernel(sig)
    Drc = (epison/math.pi)/(epison**2+LSF**2)
    Hea = 0.5*(1+(2/math.pi)*np.arctan(LSF/epison))
    Ix,Iy = np.gradient(LSF)
    s = np.sqrt(np.square(Ix)+np.square(Iy))
    Nx = Ix/(s+0.000001)
    Ny = Iy/(s+0.000001)
    Nxx,_ = np.gradient(Nx)
    _,Nyy = np.gradient(Ny)
    cur = Nxx + Nyy   
    grad_inside=Drc*(mu*cur-nu) 
    
    grad_Length=Drc*mu*cur
    
    s1 = Hea*img
    s2 = (1-Hea)*img
    s3 = 1-Hea
    C1 = s1.sum()/Hea.sum()
    C2 = s2.sum()/s3.sum()
    grad_CV_outside=Drc*(-lambda1*np.square(img-C1)+lambda2*np.square(img-C2))   
    grad_RGLS_Area=Drc*nu*abs(C1-C2)
    
    Lap=cv2.Laplacian(LSF,-1)
    grad_DR=dr*(Lap-cur)
    KIH1 = cv2.filter2D(Hea*img,-1,kernel)
    KH1 =  cv2.filter2D(Hea,-1,kernel)
    f1=KIH1/KH1
    KIH2 = cv2.filter2D((1-Hea)*img,-1,kernel) 
    KH2 = cv2.filter2D(1-Hea,-1,kernel)
    f2=KIH2/KH2
    R1 = (lambda1- lambda2)*img*img
    R2 = 2*img*cv2.filter2D(lambda1*f1 - lambda2*f2,-1,kernel)
    R3 = cv2.filter2D(lambda1*f1*f1 - lambda2*f2*f2,-1,kernel)
    grad_outside = -Drc*(R1-R2+R3) 
    
    kernel=GaussKernel(3)
    img=cv2.filter2D(img,-1,kernel)
    dx,dy=np.gradient(img)
    dI=np.sqrt(dx**2+dy**2) 

    #---------------Core code of LGMLS-------------------#  
    #grad_G
    #-------------------------End---------------------------#

    g=1/(1+dI**2)
    Tg=(g.max()+g.min())/2
    grad_RGLS_G=-kx*(g-Tg)

    if function=='LGMLS' or function=='ILLS':
        grad_LSF=grad_inside+grad_outside+grad_DR+grad_G
        LSF = LSF+step*grad_LSF 
    if function=='RSF':
        grad_LSF=grad_inside+grad_outside+grad_DR
        LSF = LSF+step*grad_LSF 
    if function=='CV':
        grad_LSF=grad_inside+grad_CV_outside
        LSF = LSF+step*grad_LSF 
    if function=='RGLS':
        grad_LSF=grad_Length+grad_RGLS_Area+grad_CV_outside+grad_RGLS_G
        LSF=LSF+step*grad_LSF       
    return LSF

def LSFtoBin(img):
    bin_img=(img>=0).astype(int)
    out_img=ndimage.binary_fill_holes(bin_img).astype(int)
    return out_img

def main(simulation,function,k,T,Gsig,filepath): 
    if simulation==False: 
        #---------------Read image-------------------#
        image=cv2.imread('./drop/'+filepath+'.tiff',1)
        img_RGB=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        #-------------------End---------------------#
    else:
        #------Generate synthesized image with PSF------#
        size=128
        gray1=0.5
        gray2=0.8

        K=GaussKernel(Gsig)
        img= np.ones((size,size))*gray1
        for i in range(size):
            for j in range(size):
                if np.sqrt(np.square(i-size/2.0)+np.square(j-size/2.0))<=32:              
                    img[i,j]=gray2          
        img=cv2.filter2D(img,-1,K)

        #Gaussian noise
        mean=0
        var=0.0002
        noise = np.random.normal(mean, var ** 0.5, img.shape)
        img = img + noise
        img = np.clip(img, 0, 1.0)
        img = np.uint8(img*255)

        # Convert to RGB
        b = np.array(img)
        g = np.array(img)
        r = np.array(img)
        img_RGB=cv2.merge([b,g,r])
         #----------------------------end--------------------------#

    plt.figure()
    plt.imshow(img,vmax=255,vmin=0,cmap='gray')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    if function=='OTSU':        
        #-------------------------OTSU---------------------------------#
        ret, th = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        print((th==255).sum())
        out=(th==255).astype(int)
        plt.figure()
        plt.imshow(out,vmax=1,vmin=0,cmap='gray')
        plt.xticks([])
        plt.yticks([])
        plt.show()
        return out,None
    
    elif function=='Threshold':
        #----------Threshold segmentation--------#
        img=255-img
        ret,th=cv2.threshold(img,T,255,cv2.THRESH_BINARY)
        print((th==255).sum())
        out=(th==255).astype(int)
        out=ndimage.binary_fill_holes(out).astype(int)
        plt.figure()
        plt.imshow(out,vmax=1,vmin=0,cmap='gray')
        plt.xticks([])
        plt.yticks([])
        plt.show()
        return out,None
    
    else:  
        #----------Level set methods,including CV,RGLS,RSF,LGMLS,and ILLS--------#
        img = np.array(img,dtype='float64')
        iniLSF=-np.ones(img.shape,img.dtype)
        iniLSF[int(0.4*img.shape[0]):int(0.6*img.shape[0]),int(0.4*img.shape[1]):int(0.6*img.shape[1])]=1
        plt.figure()
        plt.imshow(img_RGB)
        plt.contour(iniLSF, [0], linestyles ='solid',colors='r',linewidths=1.0)

        if simulation:        
            #Ground truth
            img_truth= -np.ones((size,size))
            for i in range(size):
                for j in range(size):
                    if np.sqrt(np.square(i-size/2.0)+np.square(j-size/2.0))<=32:
                        img_truth[i,j]=1
            plt.contour(img_truth, [0], linestyles ='solid',colors='g',linewidths=2.0)

        plt.xticks([])
        plt.yticks([])
        plt.show()
        
        
        mu = 0.001*255*255
        nu = 0
        lambda1=1
        lambda2=1
        dr=1
        sig=20
        num =1000
        epison = 1
        step = 0.1
        V_S=[]
        V_DS=[]
        F=0 
        
        LSF = iniLSF
        V_S.append((LSF>0).sum())
        for i in range(num):
            LSF= level_set(LSF,img,mu,nu,lambda1,lambda2,dr,sig,epison,step,k,function)
            plt.imshow(img_RGB)
            if simulation:
                plt.contour(img_truth, [0], linestyles ='solid',colors='g',linewidths=1.0)
            Bin_LSF=LSFtoBin(LSF)
            plt.contour(Bin_LSF,[0],linestyles = '-',colors='r',linewidths = 1.0)
            plt.xticks([])
            plt.yticks([])
            plt.show()
            S=(Bin_LSF>0).sum()
            V_S.append(S)
            print('演化次数:',i+1)
            print('S:',V_S[i])
        
            Diameter=Bin_LSF.sum(axis=0)
            D_max=Diameter.max()        
            print(D_max)
      
            if i>0:
                print('dS:',V_S[i]-V_S[i-1])
                V_DS.append(abs(V_S[i]-V_S[i-1]))  
                            
            if i>3:
                if np.sum(V_DS[i-t-1] for t in range(3))<=1:
                    break  

            if function=='ILLS':
                #---------------Core code of LGMLS-------------------#
                #Iterative learning
                pass
                #------------------------End------------------------#
                    
        print("finished!")
        if simulation:
            print('truth:',(img_truth>=0).sum())
        plt.figure()
        plt.title('S')
        plt.plot(V_S)
        plt.figure() 
        plt.show()
        return Bin_LSF,V_S

if __name__ == '__main__':
    Result,S=main(simulation=False,function='LGMLS',k=0.5,T=159,Gsig=3,filepath='6/1')
```
